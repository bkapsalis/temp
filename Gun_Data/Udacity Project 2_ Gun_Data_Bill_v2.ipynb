```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
gun_data_initial = pd.read_csv('gun_data.csv')
```

```python
gun_data_initial.head()
```

```python
gun_data_initial.shape
```

```python
gun_data_initial.isnull().sum()
```

```python
#Many null values so I subset for just a few meaningfull columns with few null values.
gun_data = gun_data_initial[['month', 'State', 'concealed_carry_permits', 'handgun','long_gun', 'multiple']]
```

```python
gun_data['Year-Month']= pd.to_datetime(gun_data['month']).dt.to_period('M')
gun_data.drop(columns=['month'], axis=1)
gun_data['Year']= gun_data['Year-Month'].dt.year
gun_data['Month']= gun_data['Year-Month'].dt.month
gun_data=gun_data[['Year', 'Month', 'Year-Month', 'State', 'concealed_carry_permits', 'handgun', 'long_gun',
       'multiple' ]]
```

```python
gun_data=gun_data[['Year', 'Month', 'Year-Month', 'State', 'concealed_carry_permits', 'handgun', 'long_gun',
       'multiple' ]]
```

```python
gun_data.head()
```

```python
corr_gun_df = gun_data.corr()
```

```python
import seaborn as sns
sns.set(style="ticks")

sns.pairplot(corr_gun_df)
```

```python
census_data_initial = pd.read_csv('U.S. Census Data.csv')

census_data_initial.head()
```

```python
census_data_transpose1 = census_data_initial.T.reset_index()
census_data_transpose1.rename(index={'index':'State'})
#grab true column names and set .columns 
census_data_transpose1.columns = census_data_transpose1.iloc[0]
#create a new df without the zero-th because we already set column names
census_data_transpose1=census_data_transpose1.iloc[1:]
census_data_transpose1.reset_index(inplace=True)
census_data_transpose1.reset_index(drop = True) 

census_data_transpose1.head()
```

```python
census_data_transpose1.columns
```

```python

census_data_transpose1= census_data_transpose1[['State', 'Population estimates, July 1, 2016,  (V2016)',
       'Persons under 18 years, percent, July 1, 2016,  (V2016)',
       'Persons 65 years and over, percent,  July 1, 2016,  (V2016)',
       'Female persons, percent,  July 1, 2016,  (V2016)',
       'White alone, percent, July 1, 2016,  (V2016)',
       'Black or African American alone, percent, July 1, 2016,  (V2016)',
       'American Indian and Alaska Native alone, percent, July 1, 2016,  (V2016)',
       'Asian alone, percent, July 1, 2016,  (V2016)',
       'Native Hawaiian and Other Pacific Islander alone, percent, July 1, 2016,  (V2016)',
       'Hispanic or Latino, percent, July 1, 2016,  (V2016)',
       'White alone, not Hispanic or Latino, percent, July 1, 2016,  (V2016)',
       'Veterans, 2011-2015', 'Foreign born persons, percent, 2011-2015',
       'Housing units,  July 1, 2016,  (V2016)', 'Households, 2011-2015',
       'High school graduate or higher, percent of persons age 25 years+, 2011-2015',
       "Bachelor's degree or higher, percent of persons age 25 years+, 2011-2015",
       'With a disability, under age 65 years, percent, 2011-2015',
       'Median household income (in 2015 dollars), 2011-2015',
       'Total employer establishments, 2015']]
census_data_transpose1.head()
```

```python
merged_data = pd.merge(gun_data, census_data_transpose1, on ='State', how='inner')

merged_data.shape
```

```python
sum = merged_data.groupby(['Year'], as_index=False).sum()
```

```python
sum.head()
```

```python

plt.plot(sum.Year, sum["concealed_carry_permits"], 
         c='red', label='concealed_carry_permits')
plt.plot(sum.Year, sum["handgun"], 
         c='blue', label='Handgun')

plt.xlabel('Year')

plt.ylabel('Count')
plt.title('Title')
plt.legend()
plt.show()
```
